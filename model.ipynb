**model sheshom:** 

ba tavajoh be anke natije model behtar nemishod pas bayad az model digari estefade mikardim be din jahat az **VGG-18** estefade kardam amma batavajoh be anke parametr haye ziyadi dasht ejra nemishod va error dade mishod pas **ba estefade az conv 1x1 tedad parametr hara kahesh dadam amma baz ham zaman ejra kheyli toolani bood va ejraye moodel namomken bood**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense,Dropout

#  VGG-18 
model = Sequential()

# Block 1
model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(360, 640, 3)))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

# Block 2
model.add(Conv2D(20, (1, 1), activation='relu', padding='same'))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(32, (1, 1), activation='relu', padding='same'))

model.add(MaxPooling2D((2, 2), strides=(2, 2)))

# Block 3
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(32, (1, 1), activation='relu', padding='same'))
model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model.add(Conv2D(32, (1, 1), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))


# Block 5
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(Conv2D(32, (1, 1), activation='relu'))
model.add(Conv2D(20, (3, 3), activation='relu'))
model.add(Conv2D(10, (1, 1), activation='relu'))

model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(Conv2D(15, (1, 1), activation='relu'))

# Fully connected layers
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.15))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(32, activation='relu'))  
model.add(Dense(6 ))  
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=16)
model.save('model_man-data_ziyad_charkheshi_3_royeghabli_model_jadid.keras')
```

**model haftom:**

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(360, 640, 3)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    tf.keras.layers.Conv2D(30, (1, 1), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.Conv2D(30, (1, 1), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.Conv2D(64, (1, 1), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.Conv2D(10, (1, 1), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu',),


    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.1), 
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.1),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(6 )
])
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
model.summary()
```

```python
import cv2
import numpy as np
import tensorflow as tf
import pandas as pd

model = tf.keras.models.load_model('model_man-data_ziyad_charkheshi_2_royeghabli.keras')  

def preprocess_image(image):
    img = cv2.resize(image, (640, 360))
    img = img / 255.0  
    return np.expand_dims(img, axis=0)



def draw_points(image, row):
    cv2.circle(image, (int(row['left_eye_x']), int(row['left_eye_y'])), 5, (0, 0, 255), -1)
    cv2.circle(image, (int(row['right_eye_x']), int(row['right_eye_y'])), 5, (255, 255, 0), -1)
    cv2.circle(image, (int(row['nose_x']), int(row['nose_y'])), 5, (0, 255, 255), -1)
cap = cv2.VideoCapture(0)


while cap.isOpened():
    success, image = cap.read()

    if not success:
        break

    preprocessed_image = preprocess_image(image)

    predictions =model.predict(preprocessed_image)

    df = pd.DataFrame(predictions, columns=[ 'left_eye_x', 'left_eye_y', 'right_eye_x', 'right_eye_y', 'nose_x', 'nose_y'])
    image_w=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    image_h=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    df[[ 'left_eye_x', 'right_eye_x', 'nose_x']] *= image_w

    df[[ 'left_eye_y', 'right_eye_y', 'nose_y']] *= image_h
    for index, row in df.iterrows():
        draw_points(image, row)

    cv2.imshow("Webcam Feed", cv2.flip(image,1))

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python


landmarks_model = tf.keras.models.load_model('model_man-data_kam_new_folder_1.keras')

def preprocess_image(image):
    img = cv2.resize(image, (640, 360))
    img = img / 255.0  
    return np.expand_dims(img, axis=0)

def draw_points(image, row):
    cv2.circle(image, (int(row['left_eye_x']), int(row['left_eye_y'])), 5, (0, 0, 255), -1)
    cv2.circle(image, (int(row['right_eye_x']), int(row['right_eye_y'])), 5, (255, 255, 0), -1)
    cv2.circle(image, (int(row['nose_x']), int(row['nose_y'])), 5, (0, 255, 255), -1)

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

cap = cv2.VideoCapture(0)

while cap.isOpened():
    success, image = cap.read()

    if not success:
        break

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

    preprocessed_image = preprocess_image(image)
    predictions = landmarks_model.predict(preprocessed_image)
    df = pd.DataFrame(predictions, columns=['left_eye_x', 'left_eye_y', 'right_eye_x', 'right_eye_y', 'nose_x', 'nose_y'])
    image_w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    image_h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    df[['left_eye_x', 'right_eye_x', 'nose_x']] *= image_w
    df[['left_eye_y', 'right_eye_y', 'nose_y']] *= image_h

    for index, row in df.iterrows():
        draw_points(image, row)

    cv2.imshow("Webcam Feed", cv2.flip(image, 1))

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```
